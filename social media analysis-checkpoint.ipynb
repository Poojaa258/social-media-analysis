### How platforms are performing by year in terms of user engagement in like/retweets ??????????


```python
k = df.groupby(['Platform','Year'])[['Retweets','Likes']].sum().reset_index()
k.head()
```

```python
import matplotlib.pyplot as plt

# Plotting the trend by platform
platforms = k['Platform'].unique()

plt.figure(figsize=(10, 6))

for platform in platforms:
    platform_data = k[k['Platform'] == platform]
    plt.plot(platform_data['Year'], platform_data['Likes'], marker='v', label=platform)
    
plt.xlabel('Year')
plt.ylabel('Likes')
plt.title('Trend by Platform')
plt.legend()
plt.grid(True)
plt.show()


import matplotlib.pyplot as plt

# Plotting the trend by platform
platforms = k['Platform'].unique()

plt.figure(figsize=(10, 6))


for platform in platforms:
    platform_data = k[k['Platform'] == platform]
    plt.plot(platform_data['Year'], platform_data['Retweets'], marker='v', label=platform)

plt.xlabel('Year')
plt.ylabel('Retweets')
plt.title('Trend by Platform')
plt.legend()
plt.grid(True)
plt.show()
```

### What are the things that bring  negative or positive sentiments

```python
Data_set_len = len(df) * 0.3 ######using 30%
Data_set_len
```

```python
df.head()
df.Sentiment.unique()
#Positive_Sentiments = df[df['Sentiment']== "Positive"]
g = df.groupby('Sentiment').size().sort_values(ascending = False).reset_index(name = "count")
g

g['cumsum'] = g['count'].cumsum()
g

final_df = g[g['cumsum']<= Data_set_len]
final_df
```

```python
###############Filtering a dataframe using isin command for ease
d = df[df['Sentiment'].isin(final_df.Sentiment.unique())]
d.Sentiment.unique()
```
